In this demo notebook, we will plot rewards obtained over a single training run, using a smoothing technique, to make the plot prettier.

```python
import pickle

# Unpickle the scores file
path = "../data/logs/scores.pkl"
scores = []

with open(path, "rb") as scores_file:
    scores = pickle.load(scores_file)
```

```python
import matplotlib.pyplot as plt

# Draw an unsmoothed graph
plt.plot(scores)
plt.xlabel("Episode")
plt.ylabel("Score")
```

```python
import pandas

smoothing_window = 50
smoothed_scores = (
    pandas.Series(scores).rolling(smoothing_window, min_periods=smoothing_window).mean()
)

# Plot the smoothed counterpart
plt.plot(smoothed_scores)
```
